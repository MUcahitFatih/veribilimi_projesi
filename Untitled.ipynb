```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv("data_with_nans.csv")
```

```python
data.info()
```

```python
data.head()
```

```python
data.drop(labels ="Unnamed: 0",axis=1,inplace = True)
```

```python
data.head()
```

```python
for column in data.columns[1:-1]:
    data[column].fillna(value = data[column].mean(),inplace = True)
data.isna().sum()
        
```

```python
data[data.columns[1]].plot()
```

```python
sns.scatterplot(data=data, x="Id",y="SepalLengthCm",hue ="Species")
```

```python
for column in data.columns[1:-1]:
    sns.scatterplot(data=data, x="Id",y= column,hue ="Species")
    plt.show()
    
```

```python
selected_spec = data[data["Species"]== "Iris-setosa"]
selected_column = selected_spec["SepalLengthCm"]
```

```python
std = selected_column.std()
std
```

```python
avg = selected_column.mean()
avg
```

```python
Three_sigma_plus = avg+3*std
Three_sigma_minus = avg-3*std
print(Three_sigma_plus ,Three_sigma_minus)
```

```python
selected_spec[((selected_column > Three_sigma_plus)|(selected_column < Three_sigma_minus))].index
```

```python
for column in data.columns[1:-1]:
    for spec in data["Species"].unique():
        selected_spec = data[data["Species"]== spec]
        selected_column = selected_spec[column]
        std = selected_column.std()
        avg = selected_column.mean()
        Three_sigma_plus = avg+3*std
        Three_sigma_minus = avg-3*std
        outliners=selected_spec[((selected_column > Three_sigma_plus)|(selected_column < Three_sigma_minus))].index
        data.drop(index = outliners,inplace =True)
        print(column, spec, outliners)
        
```

```python
for column in data.columns[1:-1]:
    sns.scatterplot(data=data, x="Id",y= column,hue ="Species")
    plt.show()
    
```

```python
for column in data.columns[1:-1]:
    for spec in data["Species"].unique():
        selected_spec = data[data["Species"] == spec]
        selected_column = selected_spec[column]

        q1 = selected_column.quantile(0.25)
        q3 = selected_column.quantile(0.75) 
        iqr = q3- q1 

        minimum = q1-iqr*1.5
        maximum = q3+iqr*1.5
        idxs = selected_column[(selected_column < minimum) | (selected_column>maximum)].index
          
        data.drop(index = idxs,inplace = True)
    
```

```python
for column in data.columns[1:-1]:
    sns.scatterplot(data=data, x="Id",y= column,hue ="Species")
    plt.show()
```

```python
data.to_csv("final_data.csv")
```
